```python
df = pd.read_csv(fp)
df['stop_frame']=df['stop_frame'].astype('int')
df
```

```python
sns.catplot(data=df, y = 'cycle_distance', x = 'subtech', hue='participant', aspect = 2, row='intensity', kind='box')
```

```python
a = df[df['subtech']=='dp']
b = df[df['subtech']=='dp-skate']
c = df[df['subtech']=='dp-kick']
df=pd.concat([a,b,c],axis=0)
```

```python
sns.catplot(data=df, y = 'cycle_distance', x='participant', aspect = 1.7, height=5, kind='boxen')
plt.xlabel('Participant')
plt.ylabel('Cycle Distance [m]')
plt.title('Cycle Distance - Double Pole')
```

```python
sns.catplot(data=df, y = 'cycle_time', x='participant', aspect = 1.7, height=7, kind='boxen')
plt.xlabel('Participant')
plt.ylabel('Cycle Time [s]')
plt.title('Cycle Time - Double Pole')
```
