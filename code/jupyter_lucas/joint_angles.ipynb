```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
idx = pd.IndexSlice
sns.set_style("darkgrid")
```

```python
fp = r'C:\Users\b1090197\OneDrive\Documents\XSki\xski\files\joint_angles.csv'
df = pd.read_csv(fp, header=[0,1,2,3])
df = df.iloc[:,1:]
df.columns.rename("Participant", level=0, inplace=True)
df.columns.rename("Intensity", level=1, inplace=True)
df.columns.rename("Subtech", level=2, inplace=True)
df.columns.rename("Joint", level=3, inplace=True)
df
```

```python
df.loc[:,idx[ ['P5','P8'],'hard','dp',:]]
```

```python
joints = ['Elbow', 'Shoulder', 'Hip', 'Knee']
fig, axs = plt.subplots(4,1, figsize=(10, 20))
fig
ax=axs.ravel()
for i, joint in enumerate(joints):
    ax[i].plot(df.loc[:,idx[['P5','P8'],'hard','dp','Left '+joint]], c='r', label='Left')
    ax[i].plot(df.loc[:,idx[['P5','P8'],'hard','dp','Right '+joint]], c='g', label='Right')
    ax[i].set_xlim([0,100])
    ax[i].set_xlabel('Ski Cycle [%]')
    ax[i].set_ylabel('Joint Angle [°]')
    ax[i].set_title('mean ' + joint + ' Angle (sagittal)')
    ax[i].legend()
```
