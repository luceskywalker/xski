```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
idx = pd.IndexSlice
sns.set_style("darkgrid")
```

```python
fp = r'C:\Users\b1090197\OneDrive\Documents\XSki\xski\files\mean_com_height.csv'
df = pd.read_csv(fp, header=[0,1,2])
df = df.iloc[:,1:]
df.columns.rename("Participant", level=0, inplace=True)
df.columns.rename("Intensity", level=1, inplace=True)
df.columns.rename("Subtech", level=2, inplace=True)
df
```

```python
sns.relplot(data=df.loc[:,idx[:,'hard','dp']], aspect = 2).set(
        ylabel='Height [m]',
        xlabel='Cycle [%]',
        title='COM Height - Double Pole',
        xlim = [0,100]
    )
```
