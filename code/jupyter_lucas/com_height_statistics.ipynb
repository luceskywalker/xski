```python
a = ff.loc[:,idx[:,'hard','dp']]
b = df.loc[:,idx[:,'hard','dp']]
```

```python
d = pd.concat([a,b], axis=1)
d
```

```python
p=r'C:\Users\b1090197\Documents\Case Study Kit\Recordings\data_details.xlsx'
height = pd.read_excel(p, index_col=[0])
height = height.loc['Body height'].iloc[0,1:]
height.index = a.columns.get_level_values(0).unique()
height
```

```python
c_dict={}
for i,part in enumerate(a.columns.get_level_values(0).unique()):
    com = d.values[:,[i,i+8]]*10000/height[part]
    c_dict[part]=pd.DataFrame(com, columns=['x','z'])
c_df = pd.concat(c_dict, axis=1)
c_df
```

```python
fig,ax = plt.subplots(figsize=(12,7))
for part in a.columns.get_level_values(0).unique():
    ax=sns.scatterplot(data=c_df[part], x='x', y='z',label=part)
    plt.xlabel('horizontal displacement [% BH]')
    plt.ylabel('vertical displacement [% BH]')
    plt.legend(markerscale=3, scatterpoints=1, fontsize=20)
    plt.title('Mean CoM Displacement - Double Pole - hard')
    plt.xlim([0,8])
```
